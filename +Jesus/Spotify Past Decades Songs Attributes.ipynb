```python
import pandas as pd

decades=["1950","1960","1970", "1980", "1990", "2000", "2010"]

for x in decades:
    file=f"../Datasets/Spotify Past Decades Songs Attributes/{x}.csv"
    if x=="1950":        
        df_1950=pd.read_csv(file,index_col=0)
    elif x=="1960":
        df_1960=pd.read_csv(file,index_col=0)
    elif x=="1970":
        df_1970=pd.read_csv(file,index_col=0)
    elif x=="1980":
        df_1980=pd.read_csv(file,index_col=0)
    elif x=="1990":
        df_1990=pd.read_csv(file,index_col=0)
    elif x=="2000":
        df_2000=pd.read_csv(file,index_col=0)
    elif x=="2010":
        df_2010=pd.read_csv(file,index_col=0)
```

```python
complete_dataframe=(
    df_1950.merge(df_1960, on="Number", suffixes=("_1950", "_1960"),how="outer")
    .merge(df_1970, on="Number", how="outer")
    .merge(df_1980, on="Number", suffixes=("_1970", "_1980"),how="outer")
    .merge(df_1990, on="Number",how="outer")
    .merge(df_2000, on="Number", suffixes=("_1990", "_2000"),how="outer")
    .merge(df_2010, on="Number",how="outer")
)

complete_dataframe=complete_dataframe.rename(columns={
"title":"title_2010",
"artist":"artist_2010",
"top genre":"top genre_2010",
"year":"year_2010",
"bpm":"bpm_2010",
"nrgy":"nrgy_2010",
"dnce":"dnce_2010",
"dB":"dB_2010",
"live":"live_2010",
"val":"val_2010",
"dur":"dur_2010",
"acous":"acous_2010",
"spch":"spch_2010",
"pop":"pop_2010"})

complete_dataframe.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def create_column_names(variable_to_study):  
    study_column_names=[]
    for x in range(len(decades)):
        study_column_names.append(f'{variable_to_study}_{decades[x]}')

    return study_column_names

def create_boxplots():
    result.plot(kind="bar",rot=0, figsize=(10,5),colormap='Dark2',title=variable_to_study)
#     df_to_study.plot(kind="box",rot=0, figsize=(10,5),colormap='Dark2',title=variable_to_study)
    plt.xticks(np.arange(0,7),decades)
    return plt.show()
```

```python
variable_to_study="val"

study_column_names=create_column_names(variable_to_study)
columns_for_analysis=len(study_column_names)

df_to_study=complete_dataframe[study_column_names]
result=df_to_study.mean()

create_boxplots()

# print(df_to_study)
df_to_study.describe()
```
